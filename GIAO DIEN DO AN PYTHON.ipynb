```python
#liên kết Label, button, entry
from tkinter import*
L = Tk()
L.title('Đồ án Python')
L.geometry('300x200')
L.attributes('-topmost', True)

name = Label(L, font = ('Arial', 10), text = "Họ và tên: ")
name.place(x=10, y=10)

entry = Entry(L, width = 20, font = ('Arial', 10))
entry.place (x=100,y=10)
entry.focus()

but = Button(L, text = 'Đăng nhập', width = 10, height = 2, bg = 'pink', 
             font = ('Arial', 12), fg = 'red', command = nhapvao)
but.place(x = 120, y = 70)


def nhapvao():
    name = Label(L, text = 'Bạn đã nhập: ' + entry.get(), 
                 font = ('Times New Roman', 14), bg = 'dark blue', fg = 'white')
    name.place(x = 60, y = 120)


L.mainloop()
```

```python
# Liên kết label, button, entry hoàn chỉnh
from tkinter import*
L = Tk()
L.title('Đồ án Python')
L.geometry('300x200')
L.attributes('-topmost', True)

name1 = Label(L, font = ('Arial', 10), text = "Họ và tên: ")
name1.place(x=10, y=10)

name2 = Label(L, font = ('Arial', 10), text = "Mã số SV: ")
name2.place(x=10, y=50)

entry1 = Entry(L, width = 15, font = ('Arial', 10))
entry1.place (x=100,y=10)
entry1.focus()

entry2 = Entry(L, width = 15, font = ('Arial', 10))
entry2.place (x=100,y=50)

but = Button(L, text = 'Kết quả ', width = 18, height = 1, bg = 'pink', 
             font = ('Arial', 12), fg = 'red', command = nhapvao)
but.place(x = 80, y = 90)


def nhapvao():
    a = entry1.get() + entry2.get()
    name = Label(L, text = 'Thông tin là: ' + str(a), font = ('Times New Roman', 12), bg = "blue", fg = 'white')
    name.place(x = 45, y = 130)


L.mainloop()
```
